````python
from sklearn.cluster import KMeans

k_x1 = data_class_prep.copy().drop(["PriceLevel"],axis=1)
#k_x = k_x.select_dtypes(include='number')
n_clusters = 4
cluster = KMeans(n_clusters = n_clusters,random_state=0).fit(k_x1)
y_pred = cluster.labels_
y_pred 

centroid = cluster.cluster_centers_
print(centroid.shape)
print(cluster.inertia_)
print(cluster.n_iter_)
````

````python
k_x =k_x1.copy()
k_x["pred"] = y_pred
k_x["PriceLevel"] = data_class_prep["PriceLevel"]
grouped = k_x.groupby('pred')['PriceLevel'].value_counts()
print(grouped)

k_x.loc[k_x["pred"] == 0, "pred"] = "Very High"
k_x.loc[k_x["pred"] == 1, "pred"] = "High"
k_x.loc[k_x["pred"] == 2, "pred"] = "Low"
k_x.loc[k_x["pred"] == 3, "pred"] = "Medium"

count = 0
for i in range(k_x.shape[0]):
    if k_x.loc[i, "pred"] ==k_x.loc[i,"PriceLevel"]:
        count +=1
accuracy = count/k_x.shape[0]
accuracy

# When we know the result, we can use below methods to evaluate the perform
#普通互信息分
from sklearn.metrics import adjusted_mutual_info_score
print(adjusted_mutual_info_score (k_x["pred"], k_x["PriceLevel"]) )
#调整的互信息分
from sklearn.metrics import mutual_info_score
print(mutual_info_score (k_x["pred"], k_x["PriceLevel"]) )
#标准化互信息分
from sklearn.metrics import normalized_mutual_info_score
print(normalized_mutual_info_score (k_x["pred"], k_x["PriceLevel"]))
````

````python
# when we don't know the label of the result, how to evaluate the k-means cluster
# silhouette-scores which
#```A silhouette coefficient close to 1 suggests that the clusters are 
# well-separated and distinct, with samples within each cluster tightly grouped
# together. A value of 0 indicates overlapping clusters or samples 
# on the decision boundary between clusters. Negative values usually indicate 
# that samples have been assigned to the wrong cluster or that the clusters
# are significantly overlapping.```
from sklearn.metrics import silhouette_score
from sklearn.metrics import silhouette_samples
n_clusters = 4
cluster4 = KMeans(n_clusters = n_clusters,random_state=0).fit(k_x1)
print(silhouette_score(k_x1,cluster4.labels_))
n_clusters = 7
cluster5 = KMeans(n_clusters = n_clusters,random_state=0).fit(k_x1)
print(silhouette_score(k_x1,cluster5.labels_))

# calinski_harabaz_scor other method to evaluate k-means
from sklearn.metrics import calinski_harabasz_score


print(calinski_harabasz_score(k_x1, cluster4.labels_))
print(calinski_harabasz_score(k_x1, cluster5.labels_))
````
